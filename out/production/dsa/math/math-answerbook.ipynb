## 1) Count Factors

#### Problem Statement:
    Given an integer A, you need to find the count of its factors.
    Factors of a number is the number which divides it perfectly leaving no remainder.

#### Problem Constraints:
<blockquote><span style="color:red">1</span> &le; <span style="color:green">A</span> &le; <span style="color:red">10 <sup>9</sup></span></blockquote>

#### Input Format:
    First and only argument given is an integer A
    
#### Output Format:
    Return the count of factors of A
-------------------------------------------------------------------------------------------------

#### Explanation: 
<blockquote><i>Power of observation is the key while solving any problem that comes to mind</i></blockquote>
    
    Let's take a certain example of finding the count of the factors of 32:
    
    Factors of 32:
    
    1 x 32 = 32
    2 x 16 = 32
    4 x  8 = 32
    
    So, I see that - 1, 2, 4, 8, 16, 32 are the factors of 32.
    The number of factors 32 has is 6.
    Therefore the count of factors 32 has is 6.
    

#### Solution:

    Let's find a brute force solution for this problem..
    
    Let's observe..
    
    Undoubtedly, the factors are always going to be smalller than the input integer.
    
    I can iterate through every number from 1 to 32 to check one by one sequentially, and use the modulo operator to find if the input integer is exactly divisible by any current iterator.
    Therefore, we need a for loop that iterates from 1 to N where,
<br>
<br>
    
$\forall\ x,\ from\ 1\ to\ N,\\\ when\ x\ \% N\ =\ 0,\\\ the\ count\ shall\ increment\ by 1.$ 
<br>
<br>
    
    And, if the input integer is perfectly divisible by a certain iterator, the result of the modulo operation would be zero, that means we've found one of the factors that we're looking for so we increment the count by 1.
    
    To store the count I declare a variable called 'count' and initialize it to zero, so that if there may be itertors that don't exactly divide the input integer, the count that is returned would be 0.
<br>
<br>
    
`Pseudo Code:`

```java
    count = 0;
    for(i = 1; i <= N; i++) {
        if(N % i == 0) {
            count = count + 1;
        }
    }
    return count;
```

````python
def number_of_factors(a):
    
    count = 0
    
    # this loop will run from for ranging from 1 to 32
    for i in range (1, a+1):
        if(a%i==0):
            count = count + 1
    
    return count
````
